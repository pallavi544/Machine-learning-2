```python
import pandas as pd
```

```python
df=pd.read_csv('loan.csv')
df
```

```python
df.isna().sum()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,precision_score,classification_report,confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
for column in ['gender','occupation','education_level','marital_status','income','loan_status']:
    df[column]=le.fit_transform(df[column])
df
```

```python
x=df[['age','gender','occupation','education_level','marital_status','income','credit_score']]
x
```

```python
y=df[['loan_status']]
y
```

```python
ss=StandardScaler()
ss.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
print(x_train,x_test,y_train,y_test)
```

```python
model=KNeighborsClassifier(n_neighbors=3)
model.fit(x_train,y_train)
model
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
acc=accuracy_score(y_test,y_pred)
acc
```

```python
cc=confusion_matrix(y_test,y_pred)
cc
```

```python
ccc=classification_report(y_test,y_pred)
ccc
```

```python
pp=precision_score(y_test,y_pred)
pp
```

```python
new_data=np.array([[32,1,12,1,0,24,720]])
new=ss.transform(new_data)
predicted_prob=model.predict(new)
print(predicted_prob[0])
```
